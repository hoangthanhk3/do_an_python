```python
# Vẽ đồ thị hàm số
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
x = symbols('x')
y = x**4 - 2*x**2 - 3
y_phay = diff(y,x)
print("Đạo hàm của y là:",y_phay)
y_2_phay = diff(y_phay,x)
print("Đạo hàm của y' là:",y_2_phay)
y_3_phay = diff(y_2_phay,x)
print("Đạo hàm của y'' là:",y_3_phay)
def ve_do_thi_ham_so():
  x = np.linspace(start=-10.0,stop=10.0, num=1000)
  y0 = x**4 - 2*x**2 - 3
  y1 = 4*x**3 - 4*x
  y2 = 12*x**2 - 4
  y3 = 24*x
  fig, ax = plt.subplots()
  ax.plot(x,y0,marker='x',label=r"$y = x^{4}-2x^{2}-3$")
  ax.plot(x,y1,marker='.',label=r"$y' = 4x^{3}-4x$")
  ax.plot(x,y2,marker='*',label=r"$y'' = 12x^{2}-4$")
  ax.plot(x,y3,marker='+',label=r"$y''' = 24x$")
  ax.set_xlabel("Trục hoành - x")
  ax.set_ylabel("Trục tung - y")
  ax.set_title("Đồ thị hàm số y,y',y'',y'''")
  ax.legend()
  plt.show()
# Vẽ đồ thị mặt yên ngựa
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
def do_thi_mat_yen_ngua(x,y):
  z = (x/3)**2 - (y/2)**2
  return z
def thuc_hien_ve_mat_yen_ngua():
  x = np.linspace(-10,10,500)
  y = np.linspace(-10,10,500)
  x,y = np.meshgrid(x, y)
  z = do_thi_mat_yen_ngua(x, y)
  fig,ax =plt.subplots(subplot_kw={"projection":"3d"})
  mat_yen_ngua=ax.plot_surface(x,y,z,cmap=cm.winter,linewidth=0,antialiased=False)
  plt.show()
# Vẽ đồ thị mặt hyperbolic một tầng
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
def do_thi_hyperbolic_1(x,y):
  z1 = (((x/3)**2 + (y/5)**2 - 1)*4)**1/2
  return z1
def do_thi_hyperbolic_2(x,y):
  z2 = -(((x / 3) * 2 + (y / 5) * 2 - 1) * 4) ** 1 / 2
  return z2
def thuc_hien_ve_mat_hyperbolic():
  x = np.linspace(-10, 10, 1000)
  y = np.linspace(-10, 10, 1000)
  x,y = np.meshgrid(x,y)
  z1 = do_thi_hyperbolic_1(x,y)
  z2 = do_thi_hyperbolic_2(x,y)
  fig,ax = plt.subplots(subplot_kw={"projection": "3d"})
  mat_hyperbolic = ax.plot_surface(x, y,z1,cmap=cm.cool,linewidth=0,antialiased=False)
  mat_hyperbolic = ax.plot_surface(x, y,z2,cmap=cm.cool,linewidth=0,antialiased=False)
  plt.show()
# Vẽ đồ thị mặt cầu
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
def do_thi_mat_cau1(x, y):
  z1 = 4 + (-(x + 2)**2 - (y - 1)**2 + 4)**(1 / 2)
  return z1
def do_thi_mat_cau2(x, y):
  z2 = 4 - (-(x + 2)**2 - (y - 1)**2 + 4)**(1 / 2)
  return z2
def thuc_hien_ve_mat_cau():
  x = np.linspace(-4, 0, 1500)
  y = np.linspace(-1, 3, 1500)
  x,y = np.meshgrid(x, y)
  z1 = do_thi_mat_cau1(x, y)
  z2 = do_thi_mat_cau2(x, y)
  fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
  mat_cau1 = ax.plot_surface(x,y,z1, cmap=cm.turbo, linewidth=0, antialiased=False)
  mat_cau2 = ax.plot_surface(x,y,z2, cmap=cm.turbo, linewidth=0, antialiased=False)
  plt.show()
# Chương trình chính
def main():
  thuc_hien_ve_mat_yen_ngua()
  thuc_hien_ve_mat_hyperbolic()
  thuc_hien_ve_mat_cau()
if __name__ == '__main__':
  main()




```
